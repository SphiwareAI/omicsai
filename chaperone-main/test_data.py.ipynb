```python
import numpy as np
import csv
import pandas as pd
# Define the classes and their respective probabilities
classes = {
    6.0: 'ligases',
    4.0: 'lyases',
    3.0: 'hydrolases',
    5.0: 'isomerase',
    7.0: 'translocase',
    2.0: 'transferases',
    0.0: 'non enzyme',
    1.0: 'oxidoreductase'
}

class_probabilities = [0.2, 0.1, 0.15, 0.1, 0.1, 0.15, 0.1, 0.1]

# Define the length of the protein sequences
sequence_length = 100

# Define the amino acids
amino_acids = ['A', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'Y']

# Set a random seed for reproducibility
np.random.seed(42)

# Generate synthetic protein sequences and labels
sequences = []
labels = []
for cls, class_label in classes.items():
    # Calculate the number of sequences for the current class based on the class probabilities
    num_sequences = int(class_probabilities[int(cls)] * 1000)
    
    # Generate random protein sequences for the current class
    for _ in range(num_sequences):
        sequence = ''.join(np.random.choice(amino_acids, sequence_length))
        sequences.append(sequence)
        labels.append(class_label)
```

```python
# Write the sequences and labels to a CSV file
output_file = 'protein_sequences.csv'
with open(output_file, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Protein Sequence', 'Label'])
    for sequence, label in zip(sequences, labels):
        writer.writerow([sequence, label])

print(f"Protein sequences and labels written to {output_file}.")
```
